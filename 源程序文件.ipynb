```python
#下面是西电的完整代码
#使用driver打开网址
browser3 = webdriver.Firefox()
browser3.get("https://job.xidian.edu.cn/job/search")
browser3.maximize_window()#最大化窗口

#用csv文件保存爬取数据
f = open('XIDIAN_1.csv', mode='a', encoding='utf-8-sig', newline='')
csv_writer = csv.DictWriter(f, fieldnames=['招聘主题','用人单位','发布日期','浏览次数'])
csv_writer.writeheader()
dict={}

def worm3():
    for i in range(1,245):#一共244页
        for j in range(0,20):
            job_name=browser3.find_elements(By.XPATH,"//div[@class='name']//a")
            job_time=browser3.find_elements(By.XPATH,"//div[@class='name']//span")
            job_company=browser3.find_elements(By.XPATH,"//div[@class='company']//a")
            if job_time=='2021-09-30':
                return
            dict['招聘主题']=job_name[j].text   
            dict['用人单位']=job_company[j].text      
            dict['发布日期']=job_time[j].text[:-2]
            
            
            #切换页面转换句柄
            info_button=browser3.find_elements(By.XPATH,"//div[@class='name']//a")
            info_button[0].click()#从这个网页打开新网页
            time.sleep(1)
            handles = browser3.window_handles
            browser3.switch_to.window(handles[-1])#将句柄转换
            visits=browser3.find_elements(By.XPATH,"//div[@class='share']//li")
            time.sleep(1)
            str2=visits[0].text
            str3=str2[-2:]
            dict['浏览次数']=str3
            
            csv_writer.writerow(dict)
            print(dict)
            browser3.close()#关闭当前新建页面
            time.sleep(1)
            handles = browser3.window_handles
            browser3.switch_to.window(handles[0])#把句柄重新转化回去
        next_button=browser3.find_elements(By.XPATH,"//li[@class='next']")
        next_button[0].click()#进入下一页
        time.sleep(1)
worm3()
f.close()
browser.quit()
```
